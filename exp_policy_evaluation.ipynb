This notebook is to implement policy iteration for the gridworld in Figure 4.1.

```python
import random
```

```python
N = 4

def coord2ind(i, j):
    return int(i * N + j)

def ind2coord(ind):
    return (ind // N, ind % N)
```

```python
directions = [
    (0, 1), (0, -1), (1, 0), (-1, 0)
]
p = 0.25  # random policy

v_old = [random.random() for _ in range(N * N)]
v_old[0] = v_old[-1] = 0 # terminal state

v_new = [0 for _ in range(N * N)]

def next_state(i, j, di, dj):
    ni, nj = i + di, j + dj
    if 0 <= ni and ni < N and 0 <= nj and nj < N:
        return coord2ind(ni, nj)
    return coord2ind(i, j)

while True:
    delta = 0
    for s in range(1, 15):  # not update the terminal state
        v_new[s] = 0
        i, j = ind2coord(s)
        for di, dj in directions:
            ns = next_state(i, j, di, dj)
            v_new[s] += p * (-1 + v_old[ns])
        delta = max(delta, abs(v_old[s] - v_new[s]))
    v_old = v_new[:]
    if delta < 1e-6:
        break
```

```python
grid = [[0 for _ in range (N)] for _ in range(N)]
for ind, s in enumerate(v_new):
    i, j = ind2coord(ind)
    grid[i][j] = float(s)
```

```python
import pandas as pd
from IPython.display import display

df = pd.DataFrame(grid)

styled = (
    df.style
    .format("{:.1f}") 
    .hide(axis="index")
    .hide(axis="columns")
    .set_table_styles([
        {
            "selector": "td",
            "props": [
                ("border", "1px solid black"),
                ("text-align", "center"),
                ("font-size", "18px"),
                ("width", "60px"),
                ("height", "40px"),
            ],
        },
        {
            "selector": "table",
            "props": [
                ("border-collapse", "collapse"),
                ("margin", "10px"),
            ],
        },
    ])
)

display(styled)
```
