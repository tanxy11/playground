Solve Gridworld using linear equations. 

$v_{\pi}(s) = \sum_{a}\pi(a|s) \sum_{r,s'}p(r, s'|s, a)(r + \gamma v_{\pi}(s')$

```python
import numpy as np
```

```python
N = 5

def coord2ind(i, j):
    return int(i * N + j)

def ind2coord(ind):
    return (ind // N, ind % N)
```

Environment

- (0, 1) --> (4, 1) for all $a$ with reward = 10
- (0, 3) --> (2, 3) for all $a$ with reward = 5
- if move outside the grid, stay and reward = -1
- otherwise reward = 0

```python
# set up the equation for random policy

gamma = 0.9
prob = 1.0 / 4  # random policy, p(a) = 0.25

# solve v(i, j)
A = [[0 for _ in range(N * N)] for _ in range(N * N)]
b = [0 for _ in range(N * N)]

def next_states(i, j):
    return [(i + 1, j), (i - 1, j), (i, j + 1), (i, j - 1)]

for i in range(N):
    for j in range(N):
        ind = coord2ind(i, j)
        A[ind][ind] = 1.0  # left
        if i == 0 and j == 1:
            next_ind = coord2ind(4, 1)
            b[ind] = 10
            A[ind][next_ind] -= gamma
        elif i == 0 and j == 3:
            next_ind = coord2ind(2, 3)
            b[ind] = 5
            A[ind][next_ind] -= gamma
        else:
            for ni, nj in next_states(i, j):
                if ni < 0 or nj < 0 or ni >= N or nj >= N:
                    b[ind] += -1.0 * prob
                    A[ind][ind] -= gamma * prob
                else:
                    next_ind = coord2ind(ni, nj)
                    A[ind][next_ind] -= gamma * prob
```

```python
solution = np.linalg.solve(np.array(A), np.array(b))
```

```python
grid = [[0 for _ in range (N)] for _ in range(N)]
for ind, s in enumerate(solution):
    i, j = ind2coord(ind)
    grid[i][j] = float(s)
```
